```python
import numpy as np#mathematical operations
import pandas as pd #data manipulation
```

```python
#create the data
weather_data={
    "Days":["Manday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"],
    "Temperature":[32,35,28,30,45,32,31],
    "Description":["sunny","Sunny","windy","sunny","Very sunny","Sunny","sunny"],
}
```

```python
#sort the days using temperature-from the coldest to the hottest
sorted_temperature=np.sort(weather_data["Temperature"])
print(sorted_temperature)
```


#Findings


*   The lowest temperature is 28℃
*   The highest temperature is 45℃

```python
#Display the data in a data frame
df=pd.DataFrame(weather_data)
df
```

```python
#display sorted temperature and days in a dataframe
sorted_weather=df.sort_values(by="Temperature",ascending=False)
df=pd.DataFrame(sorted_weather)
df
```

#Findings


*  The hottest day was Friday with 45℃
*   The coldest day was Wednesday with 28℃
